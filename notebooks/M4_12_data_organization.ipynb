# Milestone 4.12: Organizing Raw, Processed, and Output Artifacts

## Core Definitions
- **Raw data**: original, untouched data source (read-only)
- **Processed data**: cleaned/transformed data created from raw
- **Outputs**: final artifacts such as plots, reports, and model results

## Required Structure
```text
project/
  data/
    raw/
    processed/
  outputs/
```

## Workflow Rule
**raw -> processed -> outputs**

Keep workflow one-directional to avoid contamination and preserve reproducibility.

## Best Practices
- Never edit raw data directly
- Use clear file/folder names
- Never mix outputs inside data folders
- Keep structure self-explanatory for teammates
