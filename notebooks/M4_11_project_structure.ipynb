# Milestone 4.11: Project Folder Structure

A clean structure improves readability, collaboration, and reuse.

## Recommended Structure
```text
project/
  data/
  notebooks/
  src/
  outputs/
```

## Folder Purpose
- `data/`: raw and processed data
- `notebooks/`: exploration and narrative analysis
- `src/`: reusable Python modules
- `outputs/`: charts, reports, and exported results

## Key Rules
- Keep names simple and lowercase
- Separate data, code, and outputs
- Avoid overwriting raw data
